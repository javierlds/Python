```python
##### TERMINO DE P E M D A S   => PARENTESIS-  EXPONENTS  - MULTIPLY - DIDIVE - ADD - SUBSTRACT: Es el orden de las operaciones
# Principales operadores:
def suma(numero1,numero2):
    print('suma = ', numero1 + numero2)

def resta(numero1,numero2):
    print('resta = ', numero1 - numero2)

def multiplicacion(numero1,numero2):
    print('Multiplicacion = ', numero1 * numero2)

def division(numero1,numero2):
    print('Division = ', numero1 / numero2)
    
def potencia(numero1,numero2):
    print('Potenciacion = ', numero1 ** numero2)

def parteEnteraDiv(numero1,numero2):
    print('Parte entera de la division = ', numero1 // numero2)

def menu(seleccion):
    try:
        numero1 = int(input('Ingrese numero1:'))
        numero2 = int(input('Ingrese numero2:'))
        match seleccion:
            case 1:
                suma(numero1,numero2)
            case 2:
                resta(numero1,numero2)
            case 3:
                multiplicacion(numero1,numero2)
            case 4:
                division(numero1,numero2)
            case _:  # es el mismo default
                print('Opcion invalida')
    except ValueError:
        print('Debe indicar un valor correcto al ingresar un numero')


if __name__ == '__main__':
    try:
        while True:
            seleccion = int(input(""" Seleccione operacion
            1. Realizar suma
            2. Realizar resta
            3. Realizar multiplicacion
            4. Realizar division 
            5. Salir
            """))
            if seleccion ==5:
                break
            else:
                menu(seleccion)
    except ValueError:
        print('Debe indicar un numero valido de opcion')
   
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def odd(number):
    if number%2 != 0:
        print('Weird')

def even(number):
    if number in range(2,5):
        print('Not Weird')
    elif number in range(6,20):
        
        print('Weird')
    elif number > 20:
        print('Not Weird')
    elif number == 20:
        print('Werid')
    


if __name__ == '__main__':
    n = int(input().strip())
    if n == 0:
        odd(n)
    elif n%2 != 0:
        odd(n)
    else:
        even(n)
        
    
```

```python
#Capturar solo numero entero en una division
while True:
    try:
        numero1 = int(input("Ingrese primero numero"))
        numero2 = int(input("ingrese el numero por el cual quiere dividir"))
        total = numero1 // numero2
        print(f"el total de dividir {numero1} entre {numero2} es : {total}")
        break
    except ValueError:  #control de excepcion
        print("Por favor ingrese un numero valido")
```
