```python
##  APLICACION CON TKINTER PENDIENTE PROFUNDIZAR

import tkinter as tk
from tkinter import messagebox
import random

# Constantes para el juego
PIEDRA = 1
PAPEL = 2
TIJERA = 3

# Mapeo de opciones para visualización
opciones = {PIEDRA: 'Piedra', PAPEL: 'Papel', TIJERA: 'Tijera'}

# Ventana principal
root = tk.Tk()
root.title("Juego Piedra, Papel o Tijera")

# Variables para almacenar la elección de los jugadores
eleccion_jugador1 = None
eleccion_jugador2 = None

# Función para reiniciar el juego
def reiniciar_juego():
    global eleccion_jugador1, eleccion_jugador2
    eleccion_jugador1 = None
    eleccion_jugador2 = None
    label_resultado.config(text="¡Juega de nuevo!")

# Función para manejar la lógica del juego y mostrar el resultado
def jugar(opcion, jugador):
    global eleccion_jugador1, eleccion_jugador2
    if jugador == 1:
        eleccion_jugador1 = opcion
        label_jugador1.config(text=f"Jugador 1 eligió: {opciones[eleccion_jugador1]}")
    else:
        eleccion_jugador2 = opcion
        label_jugador2.config(text=f"Jugador 2 eligió: {opciones[eleccion_jugador2]}")

    # Si ambos jugadores han hecho su elección, determina el ganador
    if eleccion_jugador1 and eleccion_jugador2:
        resultado = [eleccion_jugador1, eleccion_jugador2]
        
        # Posibles combinaciones de victoria
        jug1Win = [[PIEDRA, TIJERA], [PAPEL, PIEDRA], [TIJERA, PAPEL]]
        jug2Win = [[TIJERA, PIEDRA], [PIEDRA, PAPEL], [PAPEL, TIJERA]]
        
        if resultado in jug1Win:
            label_resultado.config(text="Jugador 1 gana!")
        elif resultado in jug2Win:
            label_resultado.config(text="Jugador 2 gana!")
        else:
            label_resultado.config(text="¡Es un empate!")
        
        # Pregunta si desea jugar otra vez
        if messagebox.askyesno("¿Jugar de nuevo?", "¿Deseas intentarlo nuevamente?"):
            reiniciar_juego()
        else:
            root.quit()

# Configuración de interfaz
frame_jugadores = tk.Frame(root)
frame_jugadores.pack(pady=10)

label_jugador1 = tk.Label(frame_jugadores, text="Jugador 1 elige su jugada")
label_jugador1.pack(side=tk.LEFT, padx=10)

label_jugador2 = tk.Label(frame_jugadores, text="Jugador 2 elige su jugada")
label_jugador2.pack(side=tk.RIGHT, padx=10)

# Botones para elegir piedra, papel o tijera
frame_botones = tk.Frame(root)
frame_botones.pack(pady=10)

for opcion, nombre in opciones.items():
    tk.Button(frame_botones, text=nombre, command=lambda opcion=opcion: jugar(opcion, 1)).pack(side=tk.LEFT, padx=10)
    tk.Button(frame_botones, text=nombre, command=lambda opcion=opcion: jugar(opcion, 2)).pack(side=tk.RIGHT, padx=10)

# Resultado del juego
label_resultado = tk.Label(root, text="Selecciona tu jugada", font=("Helvetica", 16))
label_resultado.pack(pady=20)

# Inicia la ventana
root.mainloop()
```
