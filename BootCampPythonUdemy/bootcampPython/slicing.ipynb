```python
#Use slicing
''' Extract  parts of strings
Extructure
sequence [start: end: step]
'''
text = 'TEXTO ALTERNATIVO DE PRUEBA'
print(text[0:3])
print(text[1:8:2])
print(text[::-1])   # En esta parte muestra el texto al reves
print(text[12:4:-2])  # EN ESTA PARTE RETROCEDE DE DOS EN DOS DESDE LA POSICION 12 HACIA LA 4
```
