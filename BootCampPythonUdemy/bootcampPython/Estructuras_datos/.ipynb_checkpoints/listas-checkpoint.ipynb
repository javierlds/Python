```python
# slicing: permite sacar sublistas
lista = [30,40,60,80,240]
sublista1 = lista[1:4] # posicion 1 a la posicion 4
sublista2 = lista[:3] # hasta la posicion 3
sublista3 = lista[2:] # desde la posicion 2
sublista4 = lista[::-1] # lista invertida
print('sublista1: ' + str(sublista1))
print('sublista2: ' + str(sublista2))
print('sublista3: ' + str(sublista3))
print('sublista4:' + str(sublista4))


```

```python
# concatenacion de listas , la lista resultante muestra el orden en que se concateneron
listaprueba1 = [23,79,90]
listaprueba2 = [32,56,78,90]
listanueva = listaprueba1 + listaprueba2
for i in listanueva:
    print(i)
```

```python
#repeticion de listas,  
listaletras = ['a','b','c']
print(listaletras * 5) # el numero de veces en que se repite una lista
```

```python
#mutabilidad de listas
lista_1 = [98, '09',89]
print(lista_1)
lista_1 = ['b']
print(lista_1)
```

```python
# cambio por slicing
lista_2 = [1,3,4,5,6,78,90]
lista_2[1:3] = ['a','b']
print(lista_2)
```
