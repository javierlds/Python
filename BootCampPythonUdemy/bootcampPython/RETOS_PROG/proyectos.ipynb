```python
'''
Tienes una lista de columnas. Crea una nueva lista donde todos los nombres estén en minúsculas, sin espacios en blanco a 
los lados y los espacios intermedios se reemplacen por guiones bajos (_).
strip = elimina espacios en blanco a la derecha o izquierda, 
'''
columnas_sucias = ["  Nombre Cliente $%%$!!!-- ", "ID_USUARIO", " Fecha Nacimiento  ", "Monto Total "]
nueva_lista =  [col.upper().strip("$%!") for col in columnas_sucias] ##columnas_sucias.upper()
print(nueva_lista)
```
